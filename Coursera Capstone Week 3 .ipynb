## Question 1

```python
import urllib.request
from bs4 import BeautifulSoup
```

### Get data from website

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, "lxml")
soup.prettify()
```

```python
#Check all tables
allTables = soup.find_all("table")
allTables
```

### Get the required table

```python
#Filter and get the table
pcodeTable = soup.find('table', class_ = 'wikitable sortable')
pcodeTable
```

### Get rows from the table as a list

```python
pcodeList = []
for rows in pcodeTable.find_all('tr'):
    data = rows.find_all('td')
    rowList = []
    firstCol = 1
    for rowData in data:
        if firstCol == 1:
            rowList.append(rowData.text.rstrip())
            firstCol = 0
        else:    
            rowList.append(rowData.text.rstrip())
    pcodeList.append(rowList)
```

### Convert list to dataframe

```python
import pandas as pd

#Converting the list to a dataframe
df = pd.DataFrame(pcodeList, columns = ['PostalCode','Borough','Neighborhood'])
df.drop(df.index[0],inplace=True)
```

```python
df.head()
```

### Only process the cells that have an assigned borough. Ignore cells with a borough that is not assigned.

```python
#Get row ndices where column Borough has value "Not assigned"
dropIndices = df[ df['Borough'] =='Not assigned'].index

# Delete such rows
df.drop(dropIndices , inplace=True)
```

```python
df.head()
```

### If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
df.loc[df['Neighborhood'] =='Not assigned' , 'Neighborhood'] = df['Borough']
df.head(15)
```

### Rows will be combined into one row with the neighborhoods separated with a comma.

```python
#Join rows with same postal code area
df_grouped = df.groupby(["PostalCode", "Borough"], )["Neighborhood"].apply(", ".join).reset_index()
```

```python
#Replace '/' with ','
df_grouped['Neighborhood'] = df_grouped.Neighborhood.str.replace(' / ', ', ')
df_grouped.head()
```

```python
#Shape of the dataframe
df_grouped.shape
```

### Get upto 100 venues in each neighbourhood withing a radius of 500 metres

```python
# defining radius and limit of venues to get
radius=500
LIMIT=100
```

```python
def getNearbyVenues(names, latitudes, longitudes):
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        #API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
        #Make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        #Create a list of relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])
        
    #Create a dataframe from the list venues_list containing lists
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=df_Toronto['Neighborhood'],
                                   latitudes=df_Toronto['Latitude'],
                                   longitudes=df_Toronto['Longitude']
                                  )
toronto_venues
```

### Analyse each neighborhood

```python
# #One hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 
toronto_onehot.head()
```

```python
toronto_onehot.shape
```

### Group rows by neighborhood and by taking the mean of the frequency of occurrence of each category¶

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

### Let's print each neighborhood along with the top 5 most common venues

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

### Let's put that into a pandas dataframe

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

### Let's create a new dataframe and display the top 10 venues for each neighborhood.

```python
import numpy as np
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

### Cluster Neighborhoods

```python
#set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

#run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

#check cluster labels generated for each row in the dataframe
kmeans.labels_
```

```python
#Add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster_Labels', kmeans.labels_)

toronto_merged = df_Toronto

#Merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.head() # check the last columns!
```

```python
toronto_merged=toronto_merged.dropna() #Drop rows with unavailable data and then convert labels to integer
toronto_merged['Cluster_Labels'] = toronto_merged.Cluster_Labels.astype(int)
```

### Visualize Clusters

```python
# create map
map_clusters = folium.Map(location=[latitude_toronto, longitude_toronto], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster_Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
